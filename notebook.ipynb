```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('loanp.csv')
dfx=pd.read_csv('test.csv')
df.head()
```

```python
df.Self_Employed.unique()
```

```python
df.shape
```

```python
df1=df.drop(['Loan_ID'],axis=1)
dfx1=dfx.drop(['Loan_ID'],axis=1)
df1.head()
```

```python
df2=df1
dfx2=dfx1
df2.isnull().sum()
```

```python
df2['male'] = df1['Gender'].apply(lambda x: 1 if x=='Male' or x=='NaN' else 0)
dfx2['male'] = dfx1['Gender'].apply(lambda x: 1 if x=='Male' or x=='NaN' else 0)

df2=df2.drop(['Gender'],axis=1)
dfx2=dfx2.drop(['Gender'],axis=1)
```

```python
dfx2.head()
```

```python
location_stats = df1['Property_Area'].value_counts(ascending=False)
location_stats
```

```python
df2['Mar_y'] = df1['Married'].apply(lambda x: 1 if x=='Yes' or x=='NaN' else 0)
df2=df2.drop(['Married'],axis=1)
dfx2['Mar_y'] = dfx1['Married'].apply(lambda x: 1 if x=='Yes' or x=='NaN' else 0)
dfx2=dfx2.drop(['Married'],axis=1)
df2.head()
```

```python
df2['dep'] = df1['Dependents'].apply(lambda x: 3 if x=='3+' else x)
df2['dep'] = df2['dep'].apply(lambda x: 0 if x=='NaN' else x)
df2=df2.drop(['Dependents'],axis=1)
dfx2['dep'] = dfx1['Dependents'].apply(lambda x: 3 if x=='3+' else x)
dfx2['dep'] = dfx2['dep'].apply(lambda x: 0 if x=='NaN' else x)
dfx2=dfx2.drop(['Dependents'],axis=1)
dfx2.head()
```

```python
df2['Ed'] = df1['Education'].apply(lambda x: 1 if x=='Graduate' else 0)
df2=df2.drop(['Education'],axis=1)
dfx2['Ed'] = dfx1['Education'].apply(lambda x: 1 if x=='Graduate' else 0)
dfx2=dfx2.drop(['Education'],axis=1)
df2.head()
```

```python
dfx2.head()
```

```python
df2['semp'] = df1['Self_Employed'].apply(lambda x: 1 if x=='Yes'  else 0)
df2=df2.drop(['Self_Employed'],axis=1)
dfx2['semp'] = dfx1['Self_Employed'].apply(lambda x: 1 if x=='Yes'  else 0)
dfx2=dfx2.drop(['Self_Employed'],axis=1)
dfx2.head()
```

```python
df2.LoanAmount=df2.LoanAmount.fillna(df2.LoanAmount.mean())
dfx2.LoanAmount=dfx2.LoanAmount.fillna(dfx2.LoanAmount.mean())

df2.head()
```

```python
df2.Loan_Amount_Term=df2.Loan_Amount_Term.fillna(df2.Loan_Amount_Term.mean())
dfx2.Loan_Amount_Term=dfx2.Loan_Amount_Term.fillna(dfx2.Loan_Amount_Term.mean())

df2.head()
```

```python
df2.Credit_History=df2.Credit_History.fillna(df2.Credit_History.mean())
dfx2.Credit_History=dfx2.Credit_History.fillna(dfx2.Credit_History.mean())

df2.head()
```

```python
df2['dep'] = df2['dep'].fillna(0)
dfx2['dep'] = dfx2['dep'].fillna(0)
```

```python
ap=pd.get_dummies(df.Property_Area)
ap.drop(['Rural'],axis=1)
df2=pd.concat([df2,ap],axis=1)
apx=pd.get_dummies(dfx.Property_Area)
apx.drop(['Rural'],axis=1)
dfx2=pd.concat([dfx2,apx],axis=1)
dfx2.head()
```

```python
df2=df2.drop(['Property_Area'],axis=1)
dfx2=dfx2.drop(['Property_Area'],axis=1)

df2.head()
```

```python
y= df2.Loan_Status
y = y.apply(lambda x: 1 if x=='Y'  else 0)
```

```python
df3=df2.drop(['Loan_Status'],axis=1)
dfx3=dfx2
df3.head()
```

```python
scaler=StandardScaler()
```

```python
df3.ApplicantIncome=scaler.fit_transform(df3[['ApplicantIncome']])
dfx3.ApplicantIncome=scaler.transform(dfx3[['ApplicantIncome']])

df3.CoapplicantIncome=scaler.fit_transform(df3[['CoapplicantIncome']])
dfx3.CoapplicantIncome=scaler.transform(dfx3[['CoapplicantIncome']])

df3.LoanAmount=scaler.fit_transform(df3[['LoanAmount']])
dfx3.LoanAmount=scaler.transform(dfx3[['LoanAmount']])

df3.Loan_Amount_Term=scaler.fit_transform(df3[['Loan_Amount_Term']])
dfx3.Loan_Amount_Term=scaler.transform(dfx3[['Loan_Amount_Term']])
```

```python
df3.head()
```

```python
cross_val_score(LogisticRegression(), df3, y).mean()
```

```python
cross_val_score(DecisionTreeClassifier(), df3, y).mean()
```

```python
cross_val_score(RandomForestClassifier(n_estimators=1), df3, y).mean()
```

```python
cross_val_score(SVC(), df3, y).mean()
```

```python
model_params={
    'svm':{
        'model': SVC(gamma='auto'),
        'params': {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }
    },
    'RFC':{
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1,5,10,40,100,150]
        }
    },
    'logreg':{
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C': [1,5,10]
        } 
    }
}
```

```python
from sklearn.model_selection import GridSearchCV as gcv
scores=[]

for mn,x in model_params.items():
    clf = gcv(x['model'],x['params'],cv=5,return_train_score=False)
    clf.fit(df3,y)
    scores.append({
        'model': mn,
        'val': clf.best_score_,
        'params': clf.best_params_
    })
scores
```
